```python
def validateFloat(objNum):
    try:
        float(objNum)
        return float(objNum)
    except:
        return np.nan
```

```python
temp_series = df['TempC']
new_temp_series = temp_series.apply(validateFloat)
#have to save as a returned object after validating each value/turning to nan
```

```python
df['TempC']=new_temp_series
#saving back into the column over the old column in the dataframe
```

```python
def validateDate(objNum):
    try:
        date = pd.to_datetime(objNum)
        return date
    except:
        return np.nan
```

```python
time_series = df['CurrTimestamp']
time_series
```

```python
new_time_series=time_series.apply(validateDate)
```

```python
epoch_time = pd.to_datetime(new_time_series).astype(int) / 10**9
df['CurrTimestamp'] = epoch_time
```
